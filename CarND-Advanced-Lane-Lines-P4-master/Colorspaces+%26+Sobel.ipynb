```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = plt.imread("test_images/test5.jpg")
plt.imshow(img1)
```

```python
hls = cv2.cvtColor(img1, cv2.COLOR_RGB2HLS)
channel_names = ["H Channel", "L Channel", "S Channel"]

plt.rcParams['figure.figsize'] = (20, 25)

for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(hls[:,:,i], cmap='gray')
    plt.axis("off")
    plt.title(channel_names[i])
```

```python
plt.rcParams['figure.figsize'] = (8, 16)

l_channel = hls[:,:,1]
s_channel = hls[:,:,2]
output = np.zeros_like(l_channel)

output[(l_channel > 120) & (s_channel > 150)] = 1

plt.imshow(output, cmap='gray')
```

```python
plt.rcParams['figure.figsize'] = (8, 16)

test_image = np.zeros((960, 1280), dtype=np.uint8)
# test_image[100:600, 460:480] = 255
test_image[460:480, 450:850] = 255
plt.imshow(test_image, cmap='gray')
```

```python
sobelx = cv2.Sobel(test_image, cv2.CV_64F, 0, 1)
maxv = np.max(np.abs(sobelx))
sobelx = np.uint8(255*np.abs(sobelx)/maxv)
plt.imshow(sobelx, cmap='gray')
```
